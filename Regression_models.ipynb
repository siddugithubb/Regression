#  ML Project

## Load Data

```python
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv')
df
```

## Data preparation

### data separation as X and Y

```python
Y=df['logS']
Y
```

```python
X = df.drop('logS',axis=1)
X
```

### data splitting

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=100)
```

```python
X_train
```

```python
X_test
```

## Model Building

### Linear Regression

#### train the model

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,Y_train)
```

#### Applying the model to predict

```python
Y_train_lr_prd=lr.predict(X_train)
Y_test_lr_prd=lr.predict(X_test)
```

```python
Y_train_lr_prd
```

```python
Y_test_lr_prd
```

#### **Evaluate the Model Performance**

```python
from sklearn.metrics import mean_squared_error,r2_score
lr_train_mse=mean_squared_error(Y_train,Y_train_lr_prd)
lr_train_r2=r2_score(Y_train,Y_train_lr_prd)

lr_test_mse=mean_squared_error(Y_test,Y_test_lr_prd)
lr_test_r2=r2_score(Y_test,Y_test_lr_prd)
```

```python
print(lr_train_mse,lr_train_r2,lr_test_mse,lr_test_r2)
```

```python
lr_results=pd.DataFrame(['Linear Regression',lr_train_mse,lr_train_r2,lr_test_mse,lr_test_r2]).transpose()
lr_results
```

### Random Forest  

#### Train the Model


```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
rf.fit(X_train,Y_train)
```

#### Applying the model to make predictions

```python
Y_train_rf_prd=rf.predict(X_train)
Y_test_rf_prd=rf.predict(X_test)
```

#### Evaluate the Model performance

```python
from sklearn.metrics import mean_squared_error,r2_score
rf_train_mse=mean_squared_error(Y_train,Y_train_rf_prd)
rf_train_r2=r2_score(Y_train,Y_train_rf_prd)

rf_test_mse=mean_squared_error(Y_test,Y_test_rf_prd)
rf_test_r2=r2_score(Y_test,Y_test_rf_prd)
```

```python
rf_results=pd.DataFrame(['Random Forest',rf_train_mse,rf_train_r2,rf_test_mse,rf_test_r2]).transpose()
rf_results
```

### Model Comparision

```python
df_models=pd.concat([lr_results,rf_results],axis=0).reset_index(drop=True)
df_models
```
